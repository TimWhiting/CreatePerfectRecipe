```python

from Recipe import *
from RecipeDatabase import *
from ValidIngredients import *
import csv
recipes = RecipeDatabase()

with open('learning.csv', 'w') as csvfile:
    csvWriter = csv.writer(csvfile, delimiter=',')
    for recipe in recipes.getRecipes():
        line = []
        for input in recipes.getNormalizedInputs(recipe.getInputVectorNormalized()):
            line.append(input)
        for output in recipe.getOutputVector():
            line.append(output)
        csvWriter.writerow(line)
```
